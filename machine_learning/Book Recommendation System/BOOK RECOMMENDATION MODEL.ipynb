### Poularity based recommender which I will show on the top of my site.

### The foumula that I used to calculate the best book i.e. "(total no. of rating of each book/total no. of book)" which is the average book rating.

```python
Bookname_with_rating = Books.merge(Rating, on='ISBN')
```

```python
Number_Of_Rating = Bookname_with_rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
Number_Of_Rating.rename(columns = {'Book-Rating':'Num_Rating'},inplace = True)
```

```python
Avg_Of_Rating = Bookname_with_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
Avg_Of_Rating.rename(columns = {'Book-Rating':'Rating'}, inplace = True)
```

```python
Best_Book = Number_Of_Rating.merge(Avg_Of_Rating,on='Book-Title')
Best_Book.shape
```

```python
Best_Book = Best_Book[Best_Book['Num_Rating'] >= 250]
```

```python
Top_50_Book = Best_Book.sort_values('Rating',ascending=False).head(50)
Top_50_Book.shape
```

```python
Top_50_Book = Top_50_Book.merge(Books,on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Rating','Book-Author','Image-URL-M']]
```

### Filtering users those are given around 200 Rating On top of Books

```python
all_User_Book_Rating = Bookname_with_rating.groupby('User-ID').count()['Book-Rating'].reset_index()
all_User_Book_Rating.rename(columns={'Book-Rating':'Num_Of_User_Rating'},inplace=True)
all_User_Book_Rating
```

```python
User_Preference = all_User_Book_Rating.merge(Bookname_with_rating,on='User-ID')
```

```python
filtered_Rating = User_Preference[User_Preference['Num_Of_User_Rating'] >= 200]
```

```python
filtered_Rating = filtered_Rating.drop(columns='Num_Of_User_Rating')
```

```python
filtered_Rating.shape
```

### Filtering books those are rating under 50 atleast

```python
Valueable_Book = filtered_Rating.groupby('Book-Title').count()['Book-Rating'] >= 50
Valueable_Book = Valueable_Book[Valueable_Book].index
filtered_Books = filtered_Rating[filtered_Rating['Book-Title'].isin(Valueable_Book)]
```

```python
filtered_Books.shape
```

```python
p_Table = filtered_Books.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
p_Table.fillna(0,inplace=True)
```

```python
p_Table
```

### I put my dataset to a vector where the similarity is calculated. To achieve it I have used Cosine Similarity Function which is provided by sklearn module

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar_Books_Score = cosine_similarity(p_Table)
```

```python
similar_Books_Score.shape
```

### Below function gives the similar books of the given book

```python
def Recommend_me_a_Book(Book_name):
#      fetch the index of the book name from the Pivot table and after finding the index we can pass it to the similar array for recommendating.
    index = np.where(p_Table.index == Book_name)[0][0]
    Books_List = sorted(list(enumerate(similar_Books_Score[index])), key = lambda x:x[1], reverse = True)[1:6]
    
    required_data = []
    for x in Books_List:
        information = []
        temp_df = filtered_Books[filtered_Books['Book-Title'] == p_Table.index[x[0]]]
        information.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        information.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        information.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        required_data.apbpend(information)
    return required_data
```
